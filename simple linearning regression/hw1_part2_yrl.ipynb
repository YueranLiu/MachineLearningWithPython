```python
import matplotlib.pyplot as plt 
plt.style.use('seaborn-whitegrid') 
import numpy as np 
noise_scale = 100 
number_of_samples = 50 
x =  25*(np.random.rand(number_of_samples) - 0.8)
x = sorted(x)
x = np.array(x)
y = 1 * x**3 + 20 * x**2 + 5 * x + noise_scale*np.random.randn(number_of_samples)
parameter = np.polyfit(x, y, 3)
p = np.poly1d(parameter)
plt.plot(x,y,'ro')
plt.plot(x,p(x), color='green')
```

```python
for m in [2,3,4,5,6]:
    parameter = np.polyfit(x, y, m)
    p = np.poly1d(parameter)
    print(p)
    plt.plot(x,p(x), color='green')
```

```python
def mse(x, y, w):
    #mse = np.sum((x.dot(w) - y) ** 2) / 50  # mean square equation
    parameter = np.polyfit(x, y, w)
    p=np.polyval(parameter,x)
    mse = ((y-p) ** 2).mean()
    return mse
```

```python
for m in [1, 2, 3, 4, 5, 6, 7, 8]:
    print(mse(x,y,m))
```
